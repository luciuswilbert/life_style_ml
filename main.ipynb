Data Pre-Processing

```python
import pandas as pd

df = pd.read_csv('Final_data.csv')
display(df.info())
display(df.describe())
```

```python
corr = df.corr(numeric_only=True)['Calories_Burned'].sort_values(ascending=False)
print(corr)
```

```python
df = df.drop(columns=[col for col in df.columns if col not in ['Session_Duration (hours)', 'Experience_Level', 'Water_Intake (liters)', 'Calories_Burned']])
display(df.info())
display(df.describe())

df.to_csv('Calories_Burned.csv', index=False)
```

Analysis

```python
df.describe().T
```

```python
import matplotlib.pyplot as plt

df.hist(bins=30, figsize=(10, 6), edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df)
plt.show()

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(data=df[['Session_Duration (hours)', 'Experience_Level', 'Water_Intake (liters)']])
plt.title("Boxplot: Duration, Experience, Water Intake")
plt.show()

plt.figure(figsize=(4,5))
sns.boxplot(y=df['Calories_Burned'])
plt.title("Boxplot: Calories Burned")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(15,4))

sns.scatterplot(data=df, x='Session_Duration (hours)', y='Calories_Burned', ax=axes[0])
axes[0].set_title("Calories vs Session Duration")

sns.scatterplot(data=df, x='Experience_Level', y='Calories_Burned', ax=axes[1])
axes[1].set_title("Calories vs Experience Level")

sns.scatterplot(data=df, x='Water_Intake (liters)', y='Calories_Burned', ax=axes[2])
axes[2].set_title("Calories vs Water Intake")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15,4))

sns.regplot(data=df, x='Session_Duration (hours)', y='Calories_Burned', ax=axes[0], line_kws={'color':'red'})
axes[0].set_title("Trend: Calories vs Session Duration")

sns.regplot(data=df, x='Experience_Level', y='Calories_Burned', ax=axes[1], line_kws={'color':'red'})
axes[1].set_title("Trend: Calories vs Experience Level")

sns.regplot(data=df, x='Water_Intake (liters)', y='Calories_Burned', ax=axes[2], line_kws={'color':'red'})
axes[2].set_title("Trend: Calories vs Water Intake")

plt.tight_layout()
plt.show()
```
